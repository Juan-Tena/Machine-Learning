# Data Wrangling - La cirugía de los datos

El **data wrangling**, a veces denominada **data munging**, es el proceso de transformar y mapear datos de un dataset raw (en bruto) en otro formato con la intención de hacerlo más apropiado y valioso para una variedad de propósitos posteriores, como el análisis. Un **data wrangler** es una persona que realiza estas operaciones de transformación.

Esto puede incluir munging, visualización de datos, agregación de datos, entrenamiento de un modelo estadístico, así como muchos otros usos potenciales. La oscilación de datos como proceso generalmente sigue un conjunto de pasos generales que comienzan extrayendo los datos en forma cruda del origen de datos, dividiendo los datos en bruto usando algoritmos (por ejemplo, clasificación) o analizando los datos en estructuras de datos predefinidas, y finalmente depositando el contenido resultante en un sistema de almacenamiento (o silo) para su uso futuro.

```python
#Estos números permiten generar un solo número aleatorio cada vez.
#Si quiero generar una lista de n valores aleatorios, solo tengo que pensar en el intervalo.
def randint_list(n,a,b):
    x=[]
    for i in range(n):
        x.append(np.random.randint(a,b))#append, me permite introducir valores en una lista
    return x
```

### La simulación de Montecarlo

* Generamos dos número aleatorios, x e y, entre 0 y 1 en total 1000 veces.
* Calcularemos x*x + y*y
    * si el valor es inferior a 1, estamos dentro del círculo
    * si el valor es superior a 1, estamos fuera del círculo

* Calculamos el número total de veces que están dentro del círculo y lo dividimos entre el número total de intentos para obtener una aproximación de la probabilidad de caer dentro del círculo.

* Usamos dicha probabilidad para aproximar el valor de pi.

* Repetimos el experimento un número suficiente de veces (por ejemplo 100), para obtener (100) diferentes aproximaciones de pi.

* Calculamos el promedio de los 100 experimentos anteriores para dar un valor final de pi

```python
def pi_montecarlo(n,n_exp):
    pi_avg = 0
    pi_value_list = []

    for i in range(n_exp):
        value = 0
        x = np.random.uniform(0,1,n).tolist()
        y = np.random.uniform(0,1,n).tolist()
        for j in range(n):
            z = np.sqrt(x[j] * x[j] + y[j] * y[j])
            if z<=1:
                value += 1
        float_value=float(value)
        pi_value = float_value * 4 / n
        pi_value_list.append(pi_value)
        pi_avg += pi_value
    pi = pi_avg/n_exp

    print(pi)

    fig = plt.plot(pi_value_list)
    return (pi, fig)
```

```python
pi_montecarlo(10000,200)
```

# Agregación de datos por categoria

```python
import numpy as np
import pandas as pd
```

```python
gender = ["Male", "Female"]
income = ["Poor","Middle Class","Rich"]
```

```python
n = 500

gender_data = []
income_data = []

for i in range(0,n):
    gender_data.append(np.random.choice(gender)) #Permite asignar de manera aleatoria un valor a partir de contenido de un dataset
    income_data.append(np.random.choice(income))
```

```python
gender_data[1:10]
```

```python
income_data[1:10]
```

```python
# Z -> N(0,1)
# N(m. n) -> m + s * Z
height = 160 + 10 * np.random.randn(n)
weight = 60 + 25 * np.random.randn(n)
age = 30 + 12  * np.random.randn(n)
income = 18000 + 3500 * np.random.randn(n)
```

```python
#Se crea el dataframe a partir de un diccionario
data = pd.DataFrame(
    {
        "Gender": gender_data,
        "Economic Status": income_data,
        "Height": height.astype(int),
        "Weight": weight,
        "Age":age.astype(int), #Convertimos un array float en un array entero
        "Income": income.astype(int)
    }
)
```

```python
data.head(10)
```

```python
#Podemos hacer una agregado de las funciones que quiero y aplicarles una función
double_group.aggregate(
    {
        "Income": np.sum,
        "Age": np.mean,
        "Height": np.std
    }
)
```

```python
#En el caso de quere aplicar funciones personalizadas a un campo. utilizamos las funciones lambda
double_group.aggregate(
    {
        "Age": np.mean,
        "Height": lambda h:np.mean(h)/np.std(h)
    }
)
```

```python
#En el caso de aplicar las mismas funciones a todas las columnas
double_group.aggregate([np.sum, np.mean, np.std])
```

```python
#En el caso de aplicar una función lambda
double_group.aggregate([lambda h:np.mean(h)/np.std(h)])
```

## Transformación de variables

```python
zscore = lambda x: (x - x.mean())/x.std()
```

```python
#Tranformamos el valor de una variable aplicándole la función "transform"
z_group = double_group.transform(zscore)
z_group
```

```python
#Otra opción de usar la función transform, es utilizarla para cambiar los valores NA
#  por el resultado de una fución lambda
fill_na_mean = lambda x: x.fillna(x.mean())
```
