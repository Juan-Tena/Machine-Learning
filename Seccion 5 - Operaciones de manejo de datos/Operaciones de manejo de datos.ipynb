# Data Wrangling - La cirugía de los datos

El **data wrangling**, a veces denominada **data munging**, es el proceso de transformar y mapear datos de un dataset raw (en bruto) en otro formato con la intención de hacerlo más apropiado y valioso para una variedad de propósitos posteriores, como el análisis. Un **data wrangler** es una persona que realiza estas operaciones de transformación.

Esto puede incluir munging, visualización de datos, agregación de datos, entrenamiento de un modelo estadístico, así como muchos otros usos potenciales. La oscilación de datos como proceso generalmente sigue un conjunto de pasos generales que comienzan extrayendo los datos en forma cruda del origen de datos, dividiendo los datos en bruto usando algoritmos (por ejemplo, clasificación) o analizando los datos en estructuras de datos predefinidas, y finalmente depositando el contenido resultante en un sistema de almacenamiento (o silo) para su uso futuro.

```python
#La forma más clásica de generar un número aleatorio es entre 0 y 1 y con decimales
np.random.random()
```

```python
#Estos números permiten generar un solo número aleatorio cada vez.
#Si quiero generar una lista de n valores aleatorios, solo tengo que pensar en el intervalo.
def randint_list(n,a,b):
    x=[]
    for i in range(n):
        x.append(np.random.randint(a,b))#append, me permite introducir valores en una lista
    return x
```

```python
randint_list(25,1,50)
```

# Funciones de distribución de probabilidades

### Distribución Uniforme

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a=1 #Límite inferior
b=100 #Límite superior
n=2000000 #Número de muestras

data= np.random.uniform(a,b,n)
```

```python
%matplotlib inline
plt.hist(data)
```

### Distribución normal

```python
#La función de sitribución random.randn 
#  genera una función de distribución estandard, de media 0 y desviación estandard 1
data= np.random.randn(1000000)
```

```python
x=range(1,1000001)
plt.plot(x,data)
```

```python
plt.hist(data,250)
```

```python
#Función de distribución acumulada
plt.plot(x, sorted(data))
```

```python
#En el caso de que queramos generar una función de distribución normal
# con una media y una desviación distinta a la estandard
media=5.5
desviacion = 2.5
data = media + desviacion*np.random.randn(10000) # z = (x - media) / desviacion -> tipificar: N(0,1), X = media + desviacion * z
plt.hist(data)
```
